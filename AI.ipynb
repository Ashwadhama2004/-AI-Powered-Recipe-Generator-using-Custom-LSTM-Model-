```python
import openai

def generate_recipe(ingredients):
    prompt = f"Create a detailed recipe using these ingredients: {ingredients}."

    response = openai.ChatCompletion.create(  # Correct method
        model="gpt-3.5-turbo",
        messages=[{"role": "system", "content": prompt}]
    )

    return response["choices"][0]["message"]["content"]  # Correct response parsing

# Example usage
ingredients = input("Enter ingredients (comma-separated): ")
recipe = generate_recipe(ingredients)

print("\nüçΩÔ∏è AI-Generated Recipe:\n")
print(recipe)
```
